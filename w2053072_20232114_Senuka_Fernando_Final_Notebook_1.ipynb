```python
data = data.drop(columns=(['Patient_ID', 'Month_of_Birth', 'Occupation']))
print(data.shape)
```

#Data Cleaning

```python
#clean and standardize Sex column values
data['Sex'] = data['Sex'].replace('1', 'Male')
data['Sex'] = data['Sex'].fillna('Female')
```

```python
#fix negative tumor size values
invalid_tumor_size = data[data['Tumor_Size'] < 0]
if len(invalid_tumor_size) > 0:
    data.loc[data['Tumor_Size'] < 0, 'Tumor_Size'] = data.loc[data['Tumor_Size'] < 0, 'Tumor_Size'].abs()
```

#Handling Outliers

```python
#function to find outliers using IQR method
def find_outliers_IQR(data):
  q1=data.quantile(0.25)
  q3=data.quantile(0.75)
  IQR=q3-q1

  outliers = data[((data < (q1 - 1.5*IQR)) | (data > (q3 + 1.5*IQR)))]
  return outliers
```

```python
#check for outliers in Age variable
age_outliers = find_outliers_IQR(data['Age'])
print("\nOutliers in Age variable:")
print("Number of outliers:", len(age_outliers))
print(age_outliers)
```

```python
#handling the age values in the correct range
impossible_age_mask = (data['Age'] < 0) | (data['Age'] > 120)
impossible_age_indices = data[impossible_age_mask].index.tolist()
if len(impossible_age_indices) > 0:
    data.drop(impossible_age_indices, inplace=True)
```

```python
#check for outliers in Tumor_Size variable
tumor_size_outliers = find_outliers_IQR(data['Tumor_Size'])
print("\nOutliers in Tumor_Size variable:")
print("Number of outliers:", len(tumor_size_outliers))
print(tumor_size_outliers.head(10) if len(tumor_size_outliers) > 10 else tumor_size_outliers)
```

```python
#handle impossible tumor size values
impossible_tumor_mask = (data['Tumor_Size'] < 0) | (data['Tumor_Size'] > 500)
impossible_tumor_indices = data[impossible_tumor_mask].index.tolist()
if len(impossible_tumor_indices) > 0:
    data.drop(impossible_tumor_indices, inplace=True)
```

```python
# Filling missing Age values with mean
Mean_Age = data['Age'].mean()
data.fillna({'Age': Mean_Age}, inplace=True)
```

```python
# Dropping remaining records with missing values
data = data.dropna()
```

```python
# Standardize Mortality_Status values to ensure consistent labeling
data.loc[:, 'Mortality_Status'] = data['Mortality_Status'].replace(['Alive', 'ALIVE', 'ALive', 'alive'], 'Alive')
data.loc[:, 'Mortality_Status'] = data['Mortality_Status'].replace(['Dead', 'DEAD', 'dead'], 'Dead')
```

```python
# Encode Mortality_Status: 0 for Alive, 1 for Dead
data.loc[:, 'Mortality_Status'] = data['Mortality_Status'].map({'Alive': 0, 'Dead': 1})
```

```python
# Age and Tumor_Size to integers for appropriate data type representation
data = data.astype({"Age": "int", "Tumor_Size": "int"})
```

#Encoding Categorical Variables

```python
# Import necessary libraries for encoding
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Initialize label encoder
label_encoder = preprocessing.LabelEncoder()
```

```python
# Encode T_Stage
data['T_Stage'] = label_encoder.fit_transform(data['T_Stage'])
```

```python
# List of categorical columns to encode
categorical_cols = ['N_Stage', '6th_Stage', 'Differentiated', 'Sex', 'A_Stage',
                   'Estrogen_Status', 'Progesterone_Status']

# Encode each categorical column and store mapping
for col in categorical_cols:
    if col in data.columns and data[col].dtype == 'object':
        data[col] = label_encoder.fit_transform(data[col])
```

#Feauture Scaling

```python
# Define numerical features for scaling
numerical_features = ['Age', 'Tumor_Size', 'Regional_Node_Examined', 'Reginol_Node_Positive']
```

```python
# Checking for missing values before scaling
print(data[numerical_features].isnull().sum())
```

```python
# Fill any remaining missing values with mean
for col in numerical_features:
    if data[col].isnull().sum() > 0:
        mean_val = data[col].mean()
        data[col].fillna(mean_val, inplace=True)
```

```python
# Standardize numerical features
scaler = StandardScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```
