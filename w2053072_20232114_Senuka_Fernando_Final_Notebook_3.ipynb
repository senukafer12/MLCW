#Voting Ensemble Classifier - Mortality Status Prediction

```python
from sklearn.model_selection import train_test_split
```

```python
# intializing feature matrix and setting the target vector
X_clf = classification_df.drop('Mortality_Status', axis=1)
y_clf = classification_df['Mortality_Status']
```

```python
# Splitting the dataset into training and testing sets
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
# intializing base learners for ensemble votingClassifier
clf1 = LogisticRegression(max_iter=1000)
clf2 = KNeighborsClassifier(n_neighbors=5)
```

```python
# Creating a hard voting ensemble with Logistic Regression and KNN
ensemble_model = VotingClassifier(estimators=[
    ('lr', clf1),
    ('knn', clf2)
], voting='soft')

# Training the voting classifier on the training data
ensemble_model.fit(X_train_clf, y_train_clf)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report, RocCurveDisplay
import matplotlib.pyplot as plt
```

```python
# Predincting using ensemble model
y_pred_ensemble = ensemble_model.predict(X_test_clf)

#calculating and printing classification report of ensemble learner
print('Accuracy:', accuracy_score(y_test_clf, y_pred_ensemble))
print('Classification Report:\n', classification_report(y_test_clf, y_pred_ensemble))
```

```python
# Generating Confusion matrix
cm_ensemble_ln = confusion_matrix(y_test_clf, y_pred_ensemble, labels=ensemble_model.classes_)
disp_ensemble_ln = ConfusionMatrixDisplay(confusion_matrix=cm_ensemble_ln, display_labels=ensemble_model.classes_)
disp_ensemble_ln.plot()
plt.title('Ensemble Learner Confusion Matrix')
plt.show()
```

```python
# Visualizong the ROC Curve
RocCurveDisplay.from_estimator(ensemble_model, X_test_clf, y_test_clf)
plt.title('ROC Curve - Ensemble Learner')
plt.show()
```

#Decision Tree Regressor - Predicting Survival Months

```python
# Split features and target for regression
X_reg = regression_df.drop('Survival_Months', axis=1)
y_reg = regression_df['Survival_Months']
```

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Fully grown Decision Tree (DT-1)
dt1 = DecisionTreeRegressor()
dt1.fit(X_train_reg, y_train_reg)
y_pred_dt1 = dt1.predict(X_test_reg)
```

```python
# Pruned Decision Tree (DT-2) - max_depth = 4
dt2 = DecisionTreeRegressor(max_depth=4)
dt2.fit(X_train_reg, y_train_reg)
y_pred_dt2 = dt2.predict(X_test_reg)
```

#Evaluate Models


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Evaluating the full grown decision tree
print('DT-1 MAE:', mean_absolute_error(y_test_reg, y_pred_dt1))
print('DT-1 MSE:', mean_squared_error(y_test_reg, y_pred_dt1))
print('DT-1 R²:', r2_score(y_test_reg, y_pred_dt1))
```

```python
# Evaluating the pruned decision tree
print('DT-2 MAE:', mean_absolute_error(y_test_reg, y_pred_dt2))
print('DT-2 MSE:', mean_squared_error(y_test_reg, y_pred_dt2))
print('DT-2 R²:', r2_score(y_test_reg, y_pred_dt2))
```

```python
from sklearn.tree import plot_tree
```

```python
# Visualizing the full grown decision tree
plt.figure(figsize=(20,10))
plot_tree(dt1, feature_names=X_reg.columns, filled=True)
plt.title('Fully Grown Decision Tree (DT-1)')
plt.savefig('/content/drive/MyDrive/Machine Learning/CW/dt1_tree.svg')
plt.show()
```

```python
# Visualizing the pruned decision tree
plt.figure(figsize=(20,10))
plot_tree(dt2, feature_names=X_reg.columns, filled=True)
plt.title('Pruned Decision Tree (DT-2)')
plt.savefig('/content/drive/MyDrive/Machine Learning/CW/dt2_tree.svg')
plt.show()
```
